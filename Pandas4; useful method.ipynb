***More complex function***

```python
df["total_bill"].mean()
```

```python
def yelp(price):
    if price<10:
        return "$"
    elif price >= 10 and price <30:
        return "$$"
    else:
        return "$$$"
```

```python
df["yelp"]= df["total_bill"].apply(yelp)
```

```python
def simple(num):
    return num*3
```

```python
simple(2)
```

**used to apply a custom function to a multiple cloumn**

```python
# creating the function with two input parameters
def quality(total_bill,tip):
    if tip/total_bill > 0.25:
        return "Generous"
    else:
        return "Other"
```

```python
quality(16.99,1)
#checking with values
```

**syntex**

df[[list of columns ]].apply(lambda df: name_of_function(df[first comlumn]),df[second_column],....df[nth columnd]]),axis =1)

```python
df[["total_bill","tip"]].apply(lambda df : quality(df["total_bill"],df["tip"]),axis=1)
df[["total_bill","tip"]].apply(lambda df: quality(df["total_bill"],df["tip"]),axis=1)
```

***simplify using np.vactorize***

***Syntex***

np.vectorize(function)(dataframe columns)

```python
np.vectorize(quality)(df["total_bill"],df["tip"])
#it convert to numpy array 
```

```python
df["Quality"]= np.vectorize(quality)(df["total_bill"],df["tip"])
```
