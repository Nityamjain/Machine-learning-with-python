### Filter based on multiple condition

**using two ways**

**AND & -- where all condition must me true**

```python
df["total_bill"]>30

df['sex']== "Male"
```

```python
df[(df["total_bill"]>30) and (df['sex']== "Male")]
# here and and or  compair two value in  either side by side but here we are comparing whole series
```

```python
df[(df["total_bill"]>30) & (df['sex']== "Male")]
```

**OR | -- where either conditions are true**

```python
df[(df["total_bill"]>30) |(df['sex']== "Male")]
```

```python
df[(df["day"]=="Sun") | (df["day"]=="Sat")  | (df["day"]=="Fri")]
# it is anoying to add multiple condition for one column for multilple so we use 
```
