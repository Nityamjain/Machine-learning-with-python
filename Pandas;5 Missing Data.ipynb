### what to do with missig data

#### 1.Drop the Data

```python
help(df.drop)
```

```python
df.dropna()
# any rows wit missing value
```

```python
df.dropna(thresh=1)
# thresh : threshold argument , drop all null value rows except it has 1 not null value
```

```python
df.dropna(axis =0) 
# drop columns having  any missing value
```

```python
df.dropna(subset=["last_name"])
#only consider last_name columns
```

#### 2.Filling in Data

```python
help(df.fillna)
```

```python
df.fillna("new value")
# auto fill all null value with a give data
```

```python
df['pre_movie_score'].fillna(0)
# fill in a perticular value 
```

```python
#filling value using fuction
df['pre_movie_score'].fillna(df["pre_movie_score"].mean())
```
